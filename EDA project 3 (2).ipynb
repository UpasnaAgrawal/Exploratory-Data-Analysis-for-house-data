```python
# Replace missing values with the mean of the data of that column
dataset['GarageYrBlt'] = dataset['GarageYrBlt'].fillna(dataset['GarageYrBlt'].mean())
dataset['GarageYrBlt'] 
```

```python
#finding outlier 
```

```python
# Select numerical columns
numerical_columns = dataset.select_dtypes(include=['float64', 'int64']).columns

# Identify outliers using the IQR method
outliers_iqr = {}
for col in numerical_columns:
    Q1 = dataset[col].quantile(0.25)
    Q3 = dataset[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Find outliers
    outliers = dataset[(dataset[col] < lower_bound) | (dataset[col] > upper_bound)]
    outliers_iqr[col] = outliers[col].tolist()

print("Outliers using IQR Method:")
for col, outlier_values in outliers_iqr.items():
    print(f"{col}: {outlier_values}")
```

```python
# Using the IQR method to remove outliers
def remove_outliers_iqr(dataset, columns):
    for col in columns:
        Q1 = dataset[col].quantile(0.25)
        Q3 = dataset[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        dataset = dataset[(dataset[col] >= lower_bound) & (dataset[col] <= upper_bound)]
    return dataset

# Remove outliers from numerical columns
cleaned_data = remove_outliers_iqr(dataset, numerical_columns)
cleaned_data
```

```python
# Specify the few numerical columns to check after handle outlier
selected_removal_columns = ["LotFrontage", "TotalBsmtSF", "YearBuilt", "GarageCars" ]


# Plot boxplots for each selected column to check for remaining outliers
for col in selected_removal_columns:
    plt.figure(figsize=(10, 1))
    sns.boxplot(x=cleaned_data[col])
    plt.title(f'Boxplot of {col} after outlier handling')
    plt.xlabel(col)
    plt.show()
```
